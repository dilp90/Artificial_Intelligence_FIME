# **Tutorial for Matplotlib library**

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

**Exercise:**

Data visualization and results with Matplotlib in Python for Artificial Intelligence.

**Objective:**

Familiarize yourself with the use of Matplotlib for the visualization of data and results in artificial intelligence projects. You will learn to create basic and advanced graphs that are fundamental to analyzing the behavior of AI models.

**1. Import necessary libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

**2. Basic graphs:**

*Line graph*: Create a line graph to visualize a mathematical function, such as the sine function.

```python
x = np.linspace(0, 10, 100)  # Generates 100 points between 0 and 10
y = np.sin(x)  # Calculates the sine of each point

plt.plot(x, y, label="X sinus")  # Draws the graph/plot
plt.title("Sine plot")  # Title
plt.xlabel("x")  # Label for the x-axis
plt.ylabel("sin(x)")  # Label for the y-axis
plt.legend()  # Shows the legend
plt.grid(True)  # Shows the grid
plt.show()  # Shows the graph/plot
```

**3. Scatter plots:**

*Scatter plot*: Create a scatter plot to visualize how points are distributed in two dimensions.

```python
x = np.random.rand(100)  # Generates 100 random points in x
y = np.random.rand(100)  # Generates 100 random points in y

plt.scatter(x, y, color='blue', label="Random data points")  # Draws the scatter plot
plt.title("Scatter plot")  # Title
plt.xlabel("x")  # Label for the x-axis
plt.ylabel("y")  # Label for the y-axis
plt.legend()  # Shows the legend
plt.grid(True)  # Shows the grid
plt.show()  # Shows the graph
```

**4. Bar graphs/charts:**

*Bar chart/graph*: Visualize a bar data set, useful for comparing categories.

```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [5, 7, 3, 9, 6]

plt.bar(categorias, valores, color='orange')  # Draws the bar chart/graph
plt.title("Bar chart")  # Title
plt.xlabel("Categories")  # Label for the x-axis
plt.ylabel("Incidence")  # Label for the y-axis
plt.show()  # Shows the graph/chart
```

**5. Histogram:**

*Histogram*: Use a histogram to see the distribution of a dataset, useful for observing the distribution of features in an AI model.

```python
datos = np.random.randn(1000)  # Generates 1000 points with normal distribution

plt.hist(datos, bins=30, color='green', alpha=0.7)  # Draws the histogram
plt.title("Random data histogram")  # Title
plt.xlabel("Data")  # Label for the x-axis
plt.ylabel("Frequency")  # Label for the y-axis
plt.show()  # Shows the graph
```

**6. Subplots (Subplots):**

*Subplots (Multiple Plots)*: Shows multiple plots in the same window to compare different visualizations.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))  # Creates a row with two graphs/plots

axs[0].plot(x, y1, label="Sine of x")
axs[0].set_title("Sine plot")
axs[0].legend()

axs[1].plot(x, y2, label="Cosine of x", color='red')
axs[1].set_title("Cosine plot")
axs[1].legend()

plt.tight_layout()  # Adjusts the spacing
plt.show()  # Shows the graphs/plots
```

**7. Regression graph (if you have an AI model):**

*Regression graph*: Visualize the results of a linear regression to evaluate the performance of an AI model.

```python
# Example data for regression
X = np.random.rand(100, 1) * 10  # 100 random points between 0 and 10
y = 2 * X + 1 + np.random.randn(100, 1)  # Linear relationship with some noise

model = LinearRegression()
model.fit(X, y)  # Fits the regression model

y_pred = model.predict(X)  # Model predictions

plt.scatter(X, y, color='blue', label="Real data")  # Shows the actual data
plt.plot(X, y_pred, color='red', label="Linear regression model")  # Shows the regression line
plt.title("Linear regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

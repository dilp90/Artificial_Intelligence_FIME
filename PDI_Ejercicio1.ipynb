# **Mostrar y segmentar imagen**

En este ejercicio se cargará la imagen de una fruta con fondo negro en RGB, se convertirá a escala de grises y se segmentará el contorno de dicha fruta por medio de un umbral fijo.

```python
# Importar librerías
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargar imagen
#img = cv2.imread('/content/platano.jpg')

img = cv2.imread('/content/platano.jpg', cv2.IMREAD_COLOR)
# Convertir de BGR a RGB
imagen_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Separar canales RGB
b, g, r = cv2.split(imagen_rgb)

# Mostrar cada canal
plt.subplot(1, 3, 1)
plt.imshow(b, cmap='grey')
plt.title('Canal Azul')
plt.subplot(1, 3, 2)
plt.imshow(g, cmap='grey')
plt.title('Canal Verde')
plt.subplot(1, 3, 3)
plt.imshow(r, cmap='grey')
plt.title('Canal Rojo')
plt.show()
```

```python
#Convertir imagen RGB a escala de grises


# Convertir la imagen RGB a escala de grises
imagen_gris = cv2.cvtColor(imagen_rgb, cv2.COLOR_BGR2GRAY)

# Mostrar ambas imágenes: original y en escala de grises
plt.subplot(1, 2, 1)
plt.imshow(imagen_rgb)
plt.title('Imagen RGB')
plt.subplot(1, 2, 2)
plt.imshow(imagen_gris, cmap="grey")
plt.title('Imagen en Escala de Grises')
plt.show()

```

```python
# Segmentar una imagen con un umbral

# Aplicar un umbral fijo
umbral = 127  # Cambia este valor según el nivel deseado
_, imagen_segmentada = cv2.threshold(imagen_gris, umbral, 255, cv2.THRESH_BINARY)

# Mostrar imagen
plt.imshow(imagen_segmentada, cmap='grey')
plt.title('Imagen en Escala de Grises')
plt.show()
```
